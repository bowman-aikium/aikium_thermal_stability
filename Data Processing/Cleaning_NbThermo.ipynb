```python
import pandas as pd
import json
from IPython.display import display
```

```python
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', None)
```

```python
# Load JSON
with open("database.json", "r") as f:
    data = json.load(f)

# If it's a list of dicts:
df = pd.json_normalize(data)

# Save to CSV
df.to_csv("nbthermo.csv", index=False)
df = df[["Tm.nanoDSF.value", "Tm.DSF (SYPRO).value", "Tm.DSC.value", "Tm.Other.value", "Tm.Circular dichroism.value","Sequence.Aho.value"]]
df
```

```python
tm_cols = [
    "Tm.nanoDSF.value",
    "Tm.DSF (SYPRO).value",
    "Tm.DSC.value",
    "Tm.Other.value",
    "Tm.Circular dichroism.value"
]

# Collapse into one column
df["Tm"] = df[tm_cols].bfill(axis=1).iloc[:, 0]
```

```python
df = df.dropna(subset = ["Tm"])
cleaned = df[["Tm", "Sequence.Aho.value"]]
cleaned = cleaned.dropna()
cleaned['Sequence'] = df['Sequence.Aho.value'].str.replace('-', '', regex=False)
cleaned.drop(["Sequence.Aho.value"], inplace=True, axis=1)
cleaned
```

```python
cleaned.to_csv("NbThermo_Cleaned.csv", index = False)
```
